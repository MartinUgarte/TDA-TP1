```python
from encontrar_mejor_soborno import algoritmo_es_optimo
from contrabando_dinamica import obtener_paquetes_dinamica
from contrabando_greedy import obtener_paquetes_greedy
from constantes import VOLUMEN
import matplotlib.pyplot as plt
```

```python
tamaño_entrada = [i for i in range(0, VOLUMEN)]
```

En _magenta_ se muestran las respuestas pertenecientes al arreglo de soluciones óptimas, mientras que en _azul_ se muestran los resultados obtenidos por nuestro algoritmo _greedy_, y en _cyan_ se muestran los resultados obtenidos por nuestro algoritmo _dinámico_.

### a) Soluciones óptimas vs. Soluciones obtenidas mediante nuestro algoritmo Greedy con los datos ordenados.

```python
optimas_greedy_ordenado = []
obtenidas_greedy_ordenado = []

for n in tamaño_entrada:
    diferencias_greedy_ordenado = algoritmo_es_optimo(f"./datasets/datos{n}.csv", obtener_paquetes_greedy, True)[1]
    obtenidas_greedy_ordenado.append(diferencias_greedy_ordenado[0])
    optimas_greedy_ordenado.append(diferencias_greedy_ordenado[1])
```

```python
plt.scatter(tamaño_entrada, optimas_greedy_ordenado, c = 'm', s=80, label="Soluciones óptimas")
plt.scatter(tamaño_entrada, obtenidas_greedy_ordenado, c = 'b', s=20, label="Soluciones obtenidas")

plt.title(f'Cantidad de soluciones obtenidas vs las óptimas en Greedy ordenado')
plt.xlabel('Tamaño de entrada')
plt.ylabel('Cantidad de soluciones')
plt.legend()
plt.show()
```

Podemos observar que en su mayoría los puntos _azules_ coinciden con los _magentas_, pero no en su totalidad, esto muestra visualmente que el algoritmo _greedy_ **no** es siempre óptimo, pero que al tener la mercaderia ordenada, la mayor parte de las respuestas son óptimas.

### Soluciones óptimas vs. Soluciones obtenidas mediante un algoritmo Greedy con los datos desordenados

```python
optimas_greedy_desordenado = []
obtenidas_greedy_desordenado = []
for n in tamaño_entrada:
    diferencias_greedy_desordenado = algoritmo_es_optimo(f"./datasets/datos{n}.csv", obtener_paquetes_greedy)[1]
    obtenidas_greedy_desordenado.append(diferencias_greedy_desordenado[0])
    optimas_greedy_desordenado.append(diferencias_greedy_desordenado[1])

```

```python
plt.scatter(tamaño_entrada, optimas_greedy_desordenado, c = 'm', s=80, label="Soluciones óptimas")
plt.scatter(tamaño_entrada, obtenidas_greedy_desordenado, c = 'b', s=20, label="Soluciones obtenidas")

plt.title(f'Cantidad de soluciones obtenidas vs las óptimas en Greedy desordenado')
plt.xlabel('Tamaño de entrada')
plt.ylabel('Cantidad de soluciones')
plt.legend()
plt.show()
```

Al no recibir la mercaderia ordenada de mayor a menor, podemos ver que no solo el algoritmo _greedy_ no es óptimo, sino que en todos los casos empeora la solucion.
