```python
from encontrar_mejor_soborno import algoritmo_es_optimo
from contrabando_dinamica import obtener_paquetes_dinamica
from contrabando_greedy import obtener_paquetes_greedy
from constantes import VOLUMEN
import matplotlib.pyplot as plt
```

```python
tamaño_entrada = [i for i in range(0, VOLUMEN)]

optimas_greedy_ordenado = []
optimas_greedy_desordenado = []
optimas_dinamica_ordenado = []
optimas_dinamica_desordenado = []
obtenidas_greedy_ordenado = []
obtenidas_greedy_desordenado = []
obtenidas_dinamica_ordenado = []
obtenidas_dinamica_desordenado = []

for n in tamaño_entrada:
    diferencias_greedy_ordenado = algoritmo_es_optimo(f"./datasets/datos{n}.csv", obtener_paquetes_greedy, True)[1]
    diferencias_greedy_desordenado = algoritmo_es_optimo(f"./datasets/datos{n}.csv", obtener_paquetes_greedy)[1]
    diferencias_dinamica_ordenado = algoritmo_es_optimo(f"./datasets/datos{n}.csv", obtener_paquetes_dinamica, True)[1]
    diferencias_dinamica_desordenado = algoritmo_es_optimo(f"./datasets/datos{n}.csv", obtener_paquetes_dinamica)[1]
    obtenidas_greedy_ordenado.append(diferencias_greedy_ordenado[0])
    obtenidas_greedy_desordenado.append(diferencias_greedy_desordenado[0])
    obtenidas_dinamica_ordenado.append(diferencias_dinamica_ordenado[0])
    obtenidas_dinamica_desordenado.append(diferencias_dinamica_desordenado[0])
    optimas_greedy_ordenado.append(diferencias_greedy_ordenado[1])
    optimas_greedy_desordenado.append(diferencias_greedy_desordenado[1])
    optimas_dinamica_ordenado.append(diferencias_dinamica_ordenado[1])
    optimas_dinamica_desordenado.append(diferencias_dinamica_desordenado[1])
```

```python
plt.scatter(tamaño_entrada, optimas_greedy_ordenado, s=80, label="Soluciones óptimas")
plt.scatter(tamaño_entrada, obtenidas_greedy_ordenado, s=20, label="Soluciones obtenidas")

plt.title(f'Cantidad de soluciones obtenidas vs las óptimas en Greedy ordenado')
plt.xlabel('Tamaño de entrada')
plt.ylabel('Cantidad de soluciones')
plt.legend()
plt.show()
```

```python
plt.scatter(tamaño_entrada, optimas_greedy_desordenado, s=80, label="Soluciones óptimas")
plt.scatter(tamaño_entrada, obtenidas_greedy_desordenado, s=20, label="Soluciones obtenidas")

plt.title(f'Cantidad de soluciones obtenidas vs las óptimas en Greedy desordenado')
plt.xlabel('Tamaño de entrada')
plt.ylabel('Cantidad de soluciones')
plt.legend()
plt.show()
```

```python
plt.scatter(tamaño_entrada, optimas_dinamica_ordenado, s=80, label="Soluciones óptimas")
plt.scatter(tamaño_entrada, obtenidas_dinamica_ordenado, s=20, label="Soluciones obtenidas")

plt.title(f'Cantidad de soluciones obtenidas vs las óptimas en Dinámica ordenado')
plt.xlabel('Tamaño de entrada')
plt.ylabel('Cantidad de soluciones')
plt.legend()
plt.show()
```

```python
plt.scatter(tamaño_entrada, optimas_dinamica_desordenado, s=80, label="Soluciones óptimas")
plt.scatter(tamaño_entrada, obtenidas_dinamica_desordenado, s=20, label="Soluciones obtenidas")

plt.title(f'Cantidad de soluciones obtenidas vs las óptimas en Dinámica desordenado')
plt.xlabel('Tamaño de entrada')
plt.ylabel('Cantidad de soluciones')
plt.legend()
plt.show()
```
